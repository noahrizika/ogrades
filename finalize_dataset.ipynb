```python
# identify average duration a climb took him to complete

# at every point in comment:
# if there is a number:
# concatenate that number with a window size of [0,5] into the temp-comment column for the given climb

def appendCommentNumber(number):
    number = str(number)
    if comment.find(number) >= 0:
        subindex = comment.find(number)
        try:
            append_comment = comment[subindex : subindex + 8]
            
            # append the number with day, month, or year for context
            if append_comment.find('day') >= 0 or append_comment.find('week') >= 0 or append_comment.find('month') >= 0 or append_comment.find('year') >= 0:
                df.at[index, 'temp-comment'] = str(df.at[index, 'temp-comment']) + ', ' + append_comment
            
            # otherwise just append number without context
            else:
                df.at[index, 'temp-comment'] = str(df.at[index, 'temp-comment']) + ', ' + comment[subindex]
        except:
            df.at[index, 'temp-comment'] = str(df.at[index, 'temp-comment']) + ', ' + comment[subindex]

index = 0
for comment in df['comment']:
    if type(comment) == float:
        index += 1
        continue
    df.at[index, 'temp-comment'] = ''
    appendCommentNumber(1)
    appendCommentNumber(2)
    appendCommentNumber(3)
    appendCommentNumber(4)
    appendCommentNumber(5)
    appendCommentNumber(6)
    appendCommentNumber(7)
    appendCommentNumber(8)
    appendCommentNumber(9)
    index += 1   
```

```python

snumber_to_int = {
    'one': '1',
    'two': '2',
    'three': '3',
    'four': '4',
    'five': '5',
    'six': '6',
    'seven': '7',
    'eight': '8',
    'nine': '9',
    'ten': '10',
    'eleven': '11',
    'twelve': '12',
    'thirteen': '13',
    'fourteen': '14',
    'fifteen': '15',
    'sixteen': '16',
    'seventeen': '17',
    'eighteen': '18',
    'nineteen': '19',
    'twenty': '20',
    'first': '1',
    'second': '2',
    'third': '3',
    'fourth': '4',
    'fifth': '5',
    'sixth': '6',
    'seventh': '7',
    'eighth': '8',
    'ninth': '9',
    'tenth': '10',
    'eleventh': '11',
    'twelfth': '12',
    'thirteenth': '13',
    'fourteenth': '14',
    'fifteenth': '15',
    'sixteenth': '16',
    'seventeenth': '17',
    'eighteenth': '18',
    'nineteenth': '19',
    'twentieth': '20',
    '1st': '1',
    '2nd': '2',
    '3rd': '3',
    '4th': '4',
    '5th': '5',
    '6th': '6',
    '7th': '7',
    '8th': '8',
    '9th': '9',
    '10th': '10',
    '11th': '11',
    '12th': '12',
    '13th': '13',
    '14th': '14',
    '15th': '15',
    '16th': '16',
    '17th': '17',
    '18th': '18',
    '19th': '19',
    '20th': '20'
}

# given a comment, convert a word (based on the index of the final letter in the word) into its number (ie: two -> 2)

def getIntFromString(comment, time_ratio, substring, index):
    subindex = comment.find(substring)
    for i in range(10):
#     while subindex < len(comment):
        number_end_index = subindex - 2

        # iterate backwards to get the starting index of the word preceeding week
        number_start_index = number_end_index
        while number_start_index >= 0 and (comment[number_start_index] != ' '):
            number_start_index -= 1

        if number_start_index > 0:  
            number_start_index += 1 # index is at the space before the number
        number_end_index += 1 # index is at the number's second to last character
        snumber = comment[number_start_index : number_end_index]
        
        if snumber.find('-') >= 0:
            snumber = snumber.split('-')[0]
        
        if snumber in snumber_to_int:
            ograde = str(int(snumber_to_int[snumber]) * time_ratio)
            df.at[index, 'o-grade'] = ograde
            print(str(index) + '. if: ' + ograde + '. ' + df.at[index, 'name'])
            break

        try:
            snumber_temp = int(snumber)
            ograde = str(snumber_temp * time_ratio)
            df.at[index, 'o-grade'] = ograde
            print(str(index) + '. try: ' + ograde + '. ' + df.at[index, 'name'])
            break

        except ValueError:
#             print('ERROR: ' + str(snumber) + '. ' + df.at[index, 'name'])
            df.at[index, 'name']
            comment = comment[subindex + 1 : ]
            subindex = comment.find(substring)
```

```python
# from temp-comment to attempt-count, set o-grade

# if has X day, X week, X month or X year in it, move just that to attempt-count

index = 0
for comment in df['comment']:
    if type(comment) == float:
        index += 1
        continue        
    elif comment.find('week') >= 0:
        getIntFromString(comment, 7, 'week', index)
    elif comment.find('day') >= 0:
        getIntFromString(comment, 1, 'day', index)   
    index += 1
```

```python
nonflash_grades = ['9b', '9a+']
store_averages = []

# get the average grade of each o-graded climb for the given font grade
for actual_grade in nonflash_grades:
    sum_grades = 0
    count_grades = 0
    for index, ograde in enumerate(df['o-grade']):
        if df.at[index, 'actual-grade'] == actual_grade:
            count_grades += 1
            try:
                ograde = int(ograde)
                sum_grades += ograde
            except:
                continue
    grade_average = sum_grades/count_grades
    store_averages.append(str(round(grade_average)))

# set the average o-grade for climbs of the same font grade
for grade_index, actual_grade in enumerate(nonflash_grades):
    for index, ograde in enumerate(df['o-grade']):
        if df.at[index, 'actual-grade'] == actual_grade and pd.isnull(ograde):
            df.at[index, 'o-grade'] = store_averages[grade_index]

print(store_averages)
```
